**Question 2: Tasks to Be Performed:**
  - Create a Git working directory with feature1.txt and feature2.txt in the master branch
  - Create 3 branches develop, feature1 and feature2
  - In develop branch create develop.txt, do not stage or commit it
  - Stash this file and check out to feature1 branch
  - Create new.txt file in feature1 branch, stage and commit this file
  - Checkout to develop, unstash this file and commit
  - Please submit all the Git commands used to do the above steps

**Answer:**
  - Create a Git working directory with feature1.txt and feature2.txt in the master branch
        mkdir git-workflow-demo
        cd git-workflow-demo
        git init
        touch feature1.txt feature2.txt
        git add feature1.txt feature2.txt
        git commit -m "Add feature1.txt and feature2.txt to master branch"

  - Create 3 branches: develop, feature1, and feature2.
        git branch develop
        git branch feature1
        git branch feature2
  - In develop branch create develop.txt, do not stage or commit it.
        git checkout develop
        touch develop.txt
  - Stash this file and checkout to feature1 branch
        git stash (# Stashes the untracked develop.txt)
        git checkout feature1
  - Create new.txt file in feature1 branch, stage and commit this file.
        touch new.txt
        git add new.txt
        git commit -m "Add new.txt in feature1 branch".
  - Checkout to develop, unstash the file and commit it.
        git checkout develop
        git stash pop # Restores develop.txt
        git add develop.txt
        git commit -m "Add develop.txt to develop branch"

**3.Tasks to Be Performed:**

  - Create a Git working directory, with the following branches: Develop F1 f2
  - In the master branch, commit main.txt file
  - Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively
  - Push all these branches to GitHub
  - On local delete f2 branch
  - Delete the same branch on GitHub as well.

**Answer:-**

  1. Create a Git working directory with branches: develop, f1, f2
    - mkdir git-branch-demo
    - cd git-branch-demo
    - git init
  2. Create develop, f1, and f2 branches
    - git checkout -b develop
    - git checkout -b f1
    - git checkout -b f2
  3. In master branch, commit main.txt
    - git checkout master
    - touch main.txt
    - git add main.txt
    - git commit -m "Add main.txt in master branch"

  4. Add develop.txt to develop branch, f1.txt to f1, f2.txt to f2
    - git checkout develop
    - touch develop.txt
    - git add develop.txt
    - git commit -m "Add develop.txt in develop branch"
    - git checkout f1
    - touch f1.txt
    - git add f1.txt
    - git commit -m "Add f1.txt in f1 branch"
    - git checkout f2
    - touch f2.txt
    - git add f2.txt
    - git commit -m "Add f2.txt in f2 branch"
  5. Push all branches to GitHub
      - First, add the remote:-
        - git remote add origin https://github.com/your-username/repo-name.git
    - Push all branches
      - git push -u origin master
      - git push -u origin develop
      - git push -u origin f1
      - git push -u origin f2
    - Delete f2 branch locally
      - git branch -d f2
    - Delete f2 branch on GitHub
      - git push origin --delete f2

5. Tasks to Be Performed:
  1. Put master.txt on master branch, stage and commit
  2. Create 3 branches: public 1, public 2 and private
  3. Put public1.txt on public 1 branch, stage and commit
  4. Merge public 1 on master branch
  5. Merge public 2 on master branch
  6. Edit master.txt on private branch, stage and commit
  7. Now update branch public 1 and public 2 with new master code in private
  8. Also update new master code on master
  9. Finally update all the code on the private branch

**ANswer**:
  1. Put master.txt on master branch, stage and commit
    - mkdir git-private-public
    - cd git-private-public
    - git init
    - touch master.txt
    - git add master.txt
    - git commit -m "Add master.txt on master branch"
  2. Create 3 branches: public1, public2, and private
    - git checkout -b public1
    - git checkout master
    - git checkout -b public2
    - git checkout master
    - git checkout -b private
  3. Put public1.txt on public1 branch, stage and commit
    - git checkout public1
    - touch public1.txt
    - git add public1.txt
    - git commit -m "Add public1.txt on public1 branch"

  4. Merge public1 into master
    - git checkout master
    - git merge public1 -m "Merge public1 into master"

  5. Merge public2 into master
    - git merge public2 -m "Merge public2 into master (no changes expected yet)"

  6. Edit master.txt on private branch, stage and commit
    - git checkout private
    - echo "Updated by private branch" >> master.txt
    - git add master.txt
    - git commit -m "Edit master.txt on private branch"

  7. Update public1 and public2 with new master code from private
    - git checkout public1
    - git merge private -m "Merge private changes into public1"
    - git checkout public2
    - git merge private -m "Merge private changes into public2"

  8. Also update new master code on master (from private)
    - git checkout master
    - git merge private -m "Merge private changes into master"
  9. Finally update all the code on the private branch
    - git checkout private
    - git merge public1 -m "Merge public1 into private"
    - git merge public2 -m "Merge public2 into private"
    - git merge master -m "Merge master into private"



**5.Tasks to Be Performed:**
  1. Create a Git Flow workflow architecture on Git
  2. Create all the required branches
  3. starting from e feature branch, push the branch to the master, following the architecture
  4. Push an urgent.txt on master using hotfix

**Answer:-**

  1. Create a Git Flow Workflow Architecture
    - Git Flow is a branching model that uses the following branch types:
    - master – the production-ready code.
    - develop – integration branch for features.
    - feature/* – for new features.
    - release/* – for pre-production releases.
    - hotfix/* – for urgent fixes on master
  git init (Not required if already done)

  2. Create All Required Branches:-
    - Create develop branch from master
    - git checkout -b develop
  - Create a feature branch from develop
    - git checkout -b feature/my-feature develop

  3. Start from a Feature Branch → Push to Master

  Make some changes, stage, and commit.

    - echo "My feature code" > feature.txt
    - git add feature.txt
    - git commit -m "Add feature implementation"
  
  Merge feature branch into develop.
  - git checkout develop
  - git merge feature/my-feature

  - Create a release branch from develop.
  - git checkout -b release/v1.0

  Final tweaks (optional), then merge into master - git checkout master - git merge release/v1.0

  4.Push urgent.txt on Master Using a Hotfix.

  - git checkout master
  - git checkout -b hotfix/urgent-fix  

Add the urgent.txt file
  - echo "Urgent issue fixed" > urgent.txt
  - git add urgent.txt
  - git commit -m "Add urgent fix"

Merge hotfix into master and develop.

 - git checkout master
 - git merge hotfix/urgent-fix

 - git checkout develop
 - git merge hotfix/urgent-fix
